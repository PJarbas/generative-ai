```python
import torch
from diffusers import StableDiffusionPipeline
```

```python
pretrained_models = {
    "stabilityai/stable-diffusion-2": StableDiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-2", torch_dtype=torch.float16, revision="fp16") ,

     "runwayml/stable-diffusion-v1-5": StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16, revision="fp16"),

}
```

```python
prompt = "constellation of Andromeda"
```

```python
for model_name, pipe in pretrained_models.items():
    file_name = f"{model_name.split('/')[1]}-image.png"
    pipeline = pipe.to("cuda")
    image = pipeline(prompt, num_images_per_prompt=1, num_inference_steps=50, guidance_scale=7.5).images[0]
    image.save(file_name)
```
